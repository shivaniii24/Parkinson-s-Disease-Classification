# ***Parkinson’s Disease Classification***

Parkinson's is a disease that can cause a
nervous system disorder and affects the movement. Parkinson dataset contains biomedical measurements, 195 records of people with 23 different attributes. This data is used to differentiate healthy people and people with Parkinson’s disease.

### Build a model that can be used to differentiate healthy people from people having Parkinson’s disease.

Import our libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

Read in our dataset

```python
df= pd.read_csv("/content/pd_speech_features.csv")
```

```python
df
```

Since our data has some null values and many are close to zero.So let's replace all the null values to 0.

```python
df.replace(np.nan,0)
```

Let's check if there is any null value left.

```python
df.isnull().sum()
```

Now our data is preparted to be seperated in x and y that is input and out.

```python
x=df.iloc[:,:754]
y=df['class'].values
```

```python
x
```

```python
y
```

Now we need to train our model and then test it.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
classifier = LogisticRegression()
```

Fit the model

```python
classifier.fit(x_train, y_train)
```

Now that you have a training and testing dataset,its time to predict the values of our model

```python
pred_y = classifier.predict(x_test)
```

```python
pred_y
```

The confusion matrix is a summary of prediction results for a given classification problem. The confusion matrix shows the number of correct and incorrect predictions broken down by each class.

```python
confusion_matrix(y_test,pred_y)
```

Here everything is done so lets check the accuracy of our model.

```python
accuracy_score = accuracy_score(y_test, pred_y)
```

```python
print("accuracy = ", accuracy_score * 100, "%")
```
